## Face Detection using Haar Cascade classifer

For more details - https://docs.opencv.org/3.4/db/d28/tutorial_cascade_classifier.html

```python
# Import Libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
image = cv2.imread('images/multi_faces.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20,10))
plt.imshow(image)
```

```python
# Convert to graytscale ->for face ddetection color can be ignored as there are many other features
gray = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(20,10))
plt.imshow(gray,cmap='gray')
```

```python
# Using a pretrained Haar Classifier
face_cascade = cv2.CascadeClassifier('detector_architectures/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(gray,4,6) # Params - Image, scaleFactor, minNeighbor ->low values=>more faces 
```

```python
# Plot faces on image
img_with_detected_faces = np.copy(image)
for(x,y,w,h) in faces :
    cv2.rectangle(img_with_detected_faces,(x,y),(x+w,y+h),(255,0,0),5)
    
plt.figure(figsize=(20,10))
plt.imshow(img_with_detected_faces)
```
