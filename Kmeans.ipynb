# K-Means Clustering

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/monarch.jpg')
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
# Preparing data as per KMeans 
pixel_vals = image.reshape((-1,3))
pixel_vals = np.float32(pixel_vals)
```

```python
image.shape
```

```python
pixel_vals.shape
```

```python
# Implement KMeans Clustering

stop_criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.2)
k = 3
retval, labels, centers = cv2.kmeans(pixel_vals,k,None,stop_criteria,10,cv2.KMEANS_RANDOM_CENTERS)

centers = np.uint8(centers)
segmented_data = centers[labels.flatten()]

#reshape into original image dimensions
segmented_image = segmented_data.reshape((image.shape))
labels_reshape = labels.reshape(image.shape[0],image.shape[1])
plt.imshow(segmented_image)
```

```python
plt.imshow(labels_reshape==0,cmap='gray') # Visualising one segment
```

```python
# masking an image cluster
cluster = 0
masked_image = np.copy(image)
masked_image[labels_reshape == cluster] = [0,0,255]

plt.imshow(masked_image)
```
