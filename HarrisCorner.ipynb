## Harris Corner Detector

For more info -> https://web.archive.org/web/20210507051910/https://opencv-python-tutroals.readthedocs.io/en/latest/py_tutorials/py_feature2d/py_features_harris/py_features_harris.html

```python
# Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import cv2

%matplotlib inline

image = cv2.imread('images/waffle.jpg')
image_copy = np.copy(image)
image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
                    
```

```python
# Detect Corners
gray = cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
gray = np.float32(gray) # Required for Harris detector
dst = cv2.cornerHarris(gray,2,3,0.04) #0.04 influence number of corners detected, typical value is 0.04
dst = cv2.dilate(dst,None) # to enlarge brighter regions of the image
plt.imshow(dst,cmap='gray')
```

```python
# Identify Strong Corners
threshold = 0.01*dst.max()
corner_image = np.copy(image_copy)

for i in range(0, dst.shape[0]):
    for j in range(0, dst.shape[1]):
        if(dst[i,j]>threshold):
            cv2.circle(corner_image,(j,i),1,(0,255,0),1)
        
plt.imshow(corner_image)
```
